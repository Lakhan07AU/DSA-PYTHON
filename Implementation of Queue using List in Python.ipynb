```python
class Queue:
    def __init__(self):
        self.queue = []

    # Enqueue operation (Add element to the end of the queue)
    def enqueue(self, item):
        self.queue.append(item)
        print(f"Enqueued {item}")

    # Dequeue operation (Remove and return the element from the front of the queue)
    def dequeue(self):
        if not self.is_empty():
            return self.queue.pop(0)
        else:
            return "Queue is empty!"

    # Peek operation (Return the front element without removing it)
    def peek(self):
        if not self.is_empty():
            return self.queue[0]
        else:
            return "Queue is empty!"

    # Display operation (Print the current elements in the queue)
    def display(self):
        if not self.is_empty():
            print("Queue contents:", self.queue)
        else:
            print("Queue is empty!")

    # Check if the queue is empty
    def is_empty(self):
        return len(self.queue) == 0

# Main function to interact with the queue
def main():
    q = Queue()

    while True:
        print("\nQueue Operations:")
        print("1. Enqueue")
        print("2. Dequeue")
        print("3. Peek")
        print("4. Display")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            item = input("Enter item to enqueue: ")
            q.enqueue(item)

        elif choice == "2":
            print("Dequeued:", q.dequeue())

        elif choice == "3":
            print("Front element:", q.peek())

        elif choice == "4":
            q.display()

        elif choice == "5":
            print("Exiting the program...")
            break

        else:
            print("Invalid choice! Please choose between 1-5.")

# Run the program
if __name__ == "__main__":
    main()
```
